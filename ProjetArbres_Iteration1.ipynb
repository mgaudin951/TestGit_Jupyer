```python
import csv
import matplotlib.pyplot as plt

with open("/home/campus-3/Desktop/Projet_arbre/Data/arbres.csv") as csvfile:
    line_reader = csv.reader(csvfile)
    list_annee.pop(-1)
    list_annee.sort()
    
    anneeEnCours = 0
    listePointX = []
    listePointY = []
    counter = 0
    
    for i, row in enumerate(list_annee):
        counter += 1
        if  int(row) != anneeEnCours and anneeEnCours != 0:
            listePointX.append(counter - 1)
            counter = 1
            anneeEnCours = int(row)
            listePointY.append(row)
        if i == len(list_annee) - 1:
            listePointX.append(counter - 1)
        if anneeEnCours == 0:
            listePointY.append(row)
            anneeEnCours = int(row)
plt.bar(listePointY, listePointX)
plt.xlabel("AnnÃ©es", fontsize = 50)
plt.ylabel("Nombre d'arbres", fontsize = 50)
plt.title("Nombre d'arbres en fonction de l'annÃ©e", fontsize = 50)
plt.gcf().set_size_inches(50, 30)
plt.xticks(fontsize=40, rotation=90)
plt.yticks(fontsize=40)
```
